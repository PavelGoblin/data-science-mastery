# 14. Support Vector Machines


## 14.1 Introduction to SVM

Support Vector Machines (SVM) are a set of **supervised learning** algorithms used for classification, regression, and outlier detection.

<hr>

### Maximum Margin Concept

SVM finds a **hyperplane** that best separates classes by maximizing the margin between them. The margin is defined as the distance from the hyperplane to the nearest data points of any class.

### Support Vectors

**Support vectors** are the data points that lie closest to the hyperplane. They are critical in defining the decision boundary. If these points are removed, the position of the hyperplane would change.

<hr>


```python
# Load iris dataset to explore SVM behavior
iris = load_iris()
print ('Dataset shape: %s\n'%str(iris.data.shape))
print ('Target classes: %s\n'%iris.target_names)
```

## 14.2 Linear SVM

A linear SVM uses `sklearn.**SVC**(*kernel='linear'*)` to find a linear decision boundary that separates classes.

<hr>


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X, y = iris.data, iris.target

svm_linear = SVC(kernel='linear', random_state=42)
svm_linear.fit(X, y)
print ('Trained Linear SVM\n')
```

```python
y_pred = svm_linear.predict(X)
acc = accuracy_score(y, y_pred)
print ('Accuracy: %s\n'%acc)
print ('Support Vectors: %s\n'%len(svm_linear.support_))
```

## 14.3 RBF Kernel

The **Radial Basis Function (RBF)** kernel maps data into a higher-dimensional space, useful for non-linear data.

`sklearn.**SVC**(*kernel='rbf'*)`

<hr>


```python
svm_rbf = SVC(kernel='rbf', random_state=42)
svm_rbf.fit(X, y)
y_pred_rbf = svm_rbf.predict(X)
acc_rbf = accuracy_score(y, y_pred_rbf)
print ('RBF Accuracy: %s\n'%acc_rbf)
print ('Linear Accuracy: %s\n'%acc)
```

## 14.4 Kernel Comparison

We compare linear, poly, rbf, and sigmoid kernels on the same dataset.

<hr>


```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
accuracies = {}

for k in kernels:
    svm = SVC(kernel=k, random_state=42)
    svm.fit(X, y)
    accuracies[k] = accuracy_score(y, svm.predict(X))
print ('Kernel comparison complete\n')
```

```python
for k, a in accuracies.items():
    print ('Kernel %s: %s\n'%(k, a))
```

## 14.5 SVM Parameters: C and gamma

- **C**: Regularization parameter (higher = smaller margin, lower = larger margin)
- **gamma**: Defines how far the influence of a single training example reaches

`sklearn.**SVC**(*kernel='rbf', C=1.0, gamma='scale'*)`

<hr>


```python
C_values = [0.1, 1, 10, 100]
gamma_values = [0.001, 0.01, 0.1, 1]
best_acc = 0
best_params = {}

for C in C_values:
    for g in gamma_values:
        svm = SVC(kernel='rbf', C=C, gamma=g, random_state=42)
        svm.fit(X, y)
        acc = accuracy_score(y, svm.predict(X))
        if acc > best_acc:
            best_acc = acc
            best_params = {'C': C, 'gamma': g}
print ('Grid search complete\n')
```

```python
print ('Best Params: %s\n'%best_params)
print ('Best Accuracy: %s\n'%best_acc)
```

## 14.6 SVM for Regression (SVR)

Support Vector Regression (SVR) applies SVM principles to regression problems.

`sklearn.**SVR**(*kernel='rbf'*)`

<hr>


```python
from sklearn.svm import SVR
from sklearn.datasets import make_regression

X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=0.1, random_state=42)
svr = SVR(kernel='rbf')
svr.fit(X_reg, y_reg)
print ('SVR model trained\n')
```

```python
y_pred_reg = svr.predict(X_reg[:5])
print ('Actual: %s\n'%y_reg[:5])
print ('Predicted: %s\n'%y_pred_reg)
```
