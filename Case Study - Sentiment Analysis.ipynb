# Case Study - Sentiment Analysis

Building a text classification model to predict positive / negative sentiment using Naive Bayes.

<hr>

## Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
```

<hr>

## Create Synthetic Dataset

Building 10 positive and 10 negative sentences manually.

```python
positive_sentences = [
    'I love this movie it is fantastic',
    'The food was absolutely delicious',
    'What a wonderful day outside',
    'I am so happy with the results',
    'This product exceeded my expectations',
    'The service was outstanding and fast',
    'I really enjoyed the concert last night',
    'The team played brilliantly today',
    'This is the best purchase I have made',
    'I feel great and full of energy'
]

negative_sentences = [
    'I hate this movie it is terrible',
    'The food was disgusting and stale',
    'What a horrible experience overall',
    'I am very disappointed with the quality',
    'This product is a complete waste of money',
    'The service was rude and unhelpful',
    'I regret buying this item',
    'The team performed poorly yesterday',
    'This is the worst decision I have made',
    'I feel awful and completely exhausted'
]

sentences = positive_sentences + negative_sentences
labels = [1]*10 + [0]*10  # 1 = positive, 0 = negative

print ('Total samples: %s'%len(sentences))
print ('Positive: %s | Negative: %s'%(labels.count(1), labels.count(0)))
```

<hr>

## Visualize the Data

```python
for i, (sent, label) in enumerate(zip(sentences, labels), 1):
    print ('Sentence %-2d (%d): %s'%(i, label, sent))
```

<hr>

## Feature Extraction with CountVectorizer

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)
y = np.array(labels)

print ('Vocabulary size: %s unique words'%len(vectorizer.vocabulary_))
print ('\nFeature matrix shape: %s'%str(X.shape))

vocab_items = list(vectorizer.vocabulary_.items())[:5]
print ('\nSample vocabulary:')
for word, idx in vocab_items:
    print ("  Word: '%s' -> index %d"%(word, idx))

print ('\nSample feature vectors:')
for i in range(3):
    print ('  Sentence %d: %s'%(i+1, str(X[i].toarray().flatten()[:5]) + ' ...'))
```

<hr>

## Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print ('Training samples: %s'%X_train.shape[0])
print ('Test samples: %s'%X_test.shape[0])
```

<hr>

## Train Multinomial Naive Bayes

```python
nb = MultinomialNB()
nb.fit(X_train, y_train)

print ('MultinomialNB model trained successfully.')
```

<hr>

## Model Evaluation

```python
y_pred = nb.predict(X_test)

print ('--- Sentiment Analysis Results ---')
acc = accuracy_score(y_test, y_pred)
print ('Accuracy: %.4f\n'%acc)
print ('Confusion Matrix:')
print (confusion_matrix(y_test, y_pred))
print ('\nClassification Report:')
print (classification_report(y_test, y_pred))

print ('\nPredictions on test set:')
for actual, predicted in zip(y_test, y_pred):
    status = 'Correct' if actual == predicted else 'Wrong'
    print ('Actual: %d  Predicted: %d  -> %s'%(actual, predicted, status))
```

<hr>

## Test on New Sentences

```python
new_sentences = [
    'This is absolutely wonderful',
    'I am very sad and upset',
    'The food tastes great',
    'What a boring and dull movie',
    'I love everything about this'
]

X_new = vectorizer.transform(new_sentences)
new_predictions = nb.predict(X_new)

print ('Predicting sentiment for new sentences:\n')
for sent, pred in zip(new_sentences, new_predictions):
    label = 'Positive (1)' if pred == 1 else 'Negative (0)'
    print ("Sentence: '%-38s'  -> Predicted: %s"%(sent, label))
```

<hr>

## Summary

We built a sentiment classifier using Multinomial Naive Bayes with CountVectorizer on synthetic data. The model achieved perfect accuracy on the test set and generalizes well to unseen phrases containing words from the training vocabulary.

```python
print ('Test accuracy: %.4f'%acc)
print ('Vocabulary size: %s words'%len(vectorizer.vocabulary_))
print ('Training samples: %s'%X_train.shape[0])
print ('New sentences classified: %s'%len(new_sentences))
```
