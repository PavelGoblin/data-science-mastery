# **Project - Recommendation Engine**
<hr>


## **Overview**
This notebook builds a **content-based recommendation engine** using cosine similarity.
We generate a synthetic **user-item rating matrix** and recommend the top 5 items for any given user.
<hr>


## **Step 1: Import Libraries**
<hr>


```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
%matplotlib inline
print('Libraries imported successfully!\n')
```

## **Step 2: Generate Synthetic Data**
<hr>


```python
np.random.seed(42)
n_users, n_items = 50, 20
ratings = np.random.randint(1, 6, size=(n_users, n_items)).astype(float)
mask = np.random.rand(n_users, n_items) < 0.3
ratings[mask] = np.nan
print('Dataset generated: %d users, %d items\n'%(n_users, n_items))
print('Sparsity: %.2f%%\n'%(np.sum(mask)/(n_users*n_items)*100))
```

```python
print('Ratings shape: %s\n'%str(ratings.shape))
print('Average rating: %.2f\n'%np.nanmean(ratings))
print('Total ratings given: %d\n'%np.sum(~np.isnan(ratings)))
```

## **Step 3: Compute Cosine Similarity**
<hr>


```python
filled = np.nan_to_num(ratings, nan=0.0)
sim_matrix = cosine_similarity(filled)
print('Similarity matrix shape: %s\n'%str(sim_matrix.shape))
```

## **Step 4: Define Recommendation Function**
<hr>


```python
def recommend(user_id, ratings, sim_matrix, top_n=5):
    filled = np.nan_to_num(ratings, nan=0.0)
    user_sim = sim_matrix[user_id]
    scores = user_sim @ filled
    rated = ~np.isnan(ratings[user_id])
    scores[rated] = -np.inf
    top_items = np.argsort(scores)[-top_n:][::-1]
    return top_items, scores[top_items]

print('Recommendation function ready.\n')
```

## **Step 5: Get Top-5 Recommendations for User 0**
<hr>


```python
user_id = 0
items, scores = recommend(user_id, ratings, sim_matrix)
print('Top 5 recommendations for User %d:\n'%user_id)
for i, (item, score) in enumerate(zip(items, scores)):
    print('  %d. Item %d (predicted: %.2f)\n'%(i+1, item, score))
```

## **Step 6: Recommend for Multiple Users**
<hr>


```python
print('Recommendations for first 5 users:\n')
for uid in range(5):
    items, scores = recommend(uid, ratings, sim_matrix)
    line = 'User %d: '%uid
    for item, score in zip(items, scores):
        line += 'Item %d(%.2f) '%(item, score)
    print(line + '\n')
```

## **Step 7: Evaluation**
<hr>
We evaluate using **Prediction Coverage** and **Average Predicted Rating**.


```python
all_scores = []
all_items = set()
for uid in range(n_users):
    items, scores = recommend(uid, ratings, sim_matrix)
    all_scores.extend(scores)
    all_items.update(items)
print('Evaluation complete for %d users.\n'%n_users)
print('Average predicted score across all recommendations: %.2f\n'%np.mean(all_scores))
print('Total unique items recommended: %d\n'%len(all_items))
```

## **Step 8: Visualize Rating Distribution**
<hr>


```python
plt.figure(figsize=(8, 4))
plt.hist(ratings[~np.isnan(ratings)], bins=5, edgecolor='black', alpha=0.7)
plt.title('Distribution of User Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## **Conclusion**
<hr>
The **content-based recommendation engine** successfully generates personalized top-5 item recommendations.
Future improvements could include **collaborative filtering** and **matrix factorization**.
