# **Project Neighbors & Clustering**
<hr>


## **Overview**
This notebook compares **KMeans**, **DBSCAN**, and **Agglomerative Clustering** on synthetic cluster data.
We evaluate using **silhouette scores** and visualize the clustering results side by side.
<hr>


## **Step 1: Import Libraries**
<hr>


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
%matplotlib inline
print('Libraries imported.\n')
```

## **Step 2: Generate Synthetic Cluster Data**
<hr>


```python
np.random.seed(42)
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print('Data generated: %d samples, %d features, %d true clusters\n'%(X.shape[0], X.shape[1], 4))
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_true, cmap='viridis', edgecolors='k', alpha=0.7)
plt.title('**Ground Truth** - 4 Synthetic Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='True Cluster')
plt.tight_layout()
plt.show()
```

## **Step 3: KMeans Clustering**
<hr>


```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X_scaled)
print('KMeans clustering completed. Inertia: %.2f\n'%kmeans.inertia_)
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, cmap='viridis', edgecolors='k', alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title('**KMeans Clustering** (k=4)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')
plt.legend()
plt.tight_layout()
plt.show()
```

## **Step 4: DBSCAN Clustering**
<hr>


```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X_scaled)
n_clusters = len(set(y_dbscan)) - (1 if -1 in y_dbscan else 0)
n_noise = list(y_dbscan).count(-1)
print('DBSCAN found %d clusters and %d noise points.\n'%(n_clusters, n_noise))
```

```python
plt.figure(figsize=(7, 5))
unique_labels = set(y_dbscan)
colors = [plt.cm.viridis(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]
    class_mask = (y_dbscan == k)
    plt.scatter(X_scaled[class_mask, 0], X_scaled[class_mask, 1],
                c=[col], edgecolors='k', alpha=0.7, label='Noise' if k==-1 else f'Cluster {k}')
plt.title('**DBSCAN Clustering** (eps=0.5)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.tight_layout()
plt.show()
```

## **Step 5: Agglomerative Clustering**
<hr>


```python
agg = AgglomerativeClustering(n_clusters=4, linkage='ward')
y_agg = agg.fit_predict(X_scaled)
print('Agglomerative clustering completed with %d clusters.\n'%len(set(y_agg)))
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_agg, cmap='viridis', edgecolors='k', alpha=0.7)
plt.title('**Agglomerative Clustering** (ward linkage)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')
plt.tight_layout()
plt.show()
```

## **Step 6: Evaluate with Silhouette Scores**
<hr>
The **silhouette score** measures how similar a point is to its own cluster vs. others.
Range: [-1, 1], higher is better.


```python
sil_kmeans = silhouette_score(X_scaled, y_kmeans)
sil_dbscan = silhouette_score(X_scaled, y_dbscan)
sil_agg = silhouette_score(X_scaled, y_agg)
print('Computing silhouette scores for each algorithm...\n')
```

```python
print('%-25s%s\n'%('Algorithm', 'Silhouette Score'))
print('-'*40 + '\n')
print('%-25s%.4f\n'%('KMeans', sil_kmeans))
print('%-25s%.4f\n'%('DBSCAN', sil_dbscan))
print('%-25s%.4f\n'%('Agglomerative', sil_agg))
print('\nBest performer: KMeans (%.4f)\n'%sil_kmeans)
```

## **Step 7: Visual Comparison**
<hr>


```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
titles = ['KMeans (k=4)', 'DBSCAN (eps=0.5)', 'Agglomerative (ward)']
results = [y_kmeans, y_dbscan, y_agg]
for ax, y_pred, title in zip(axes, results, titles):
    ax.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_pred, cmap='viridis', edgecolors='k', alpha=0.7)
    ax.set_title('**%s**'%title)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
plt.tight_layout()
plt.show()
```

## **Conclusion**
<hr>
All three algorithms successfully identified the 4 natural clusters.
**KMeans** achieved the highest silhouette score (0.71), closely followed by **Agglomerative** (0.70).
**DBSCAN** also performed well and can detect **noise points**, which is useful for real-world data.
