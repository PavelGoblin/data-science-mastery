# Sony - Pandas & Basic Visualization

<hr>
**DataFrame operations + matplotlib plots**
<hr>

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print ('imports done\n')
```

<hr>
**1. Create DataFrame**
<hr>

```python
np.random.seed(42)
df = pd.DataFrame({
    'name': ['Alice','Bob','Charlie','David','Eve']*20,
    'dept': np.random.choice(['IT','HR','Sales','Finance'], size=100),
    'salary': np.random.randint(30000, 120000, size=100),
    'experience': np.random.randint(1, 25, size=100),
    'score': np.random.uniform(60, 100, size=100).round(1)
})
print ('shape: %s\n'%str(df.shape))
print (df.head())
```

```python
print (df.info())
print ('\n')
print (df.describe())
```

<hr>
**2. Selecting and filtering**
<hr>

```python
print ('first 3 rows:\n%s\n'%str(df[:3]))
print ('salary column type: %s\n'%str(type(df['salary'])))
print ('IT dept rows: %d\n'%len(df[df['dept']=='IT']))
print ('salary > 90000:\n%s\n'%str(df[df['salary']>90000][['name','salary']].head()))
print ('high score & experience:\n%s\n'%str(df[(df['score']>90)&(df['experience']>10)].head()))
```

<hr>
**3. GroupBy and aggregation**
<hr>

```python
dept_stats = df.groupby('dept').agg({
    'salary': ['mean','std','count'],
    'experience': 'mean',
    'score': 'mean'
}).round(2)
print ('dept stats:\n%s\n'%str(dept_stats))
```

```python
name_stats = df.groupby('name')['salary'].agg(['mean','min','max','std']).round(2)
print ('name stats:\n%s\n'%str(name_stats))
```

<hr>
**4. Handling missing data**
<hr>

```python
df_missing = df.copy()
df_missing.loc[np.random.choice(100, 10), 'salary'] = np.nan
df_missing.loc[np.random.choice(100, 5), 'score'] = np.nan
print ('missing values:\n%s\n'%str(df_missing.isnull().sum()))
print ('fill with median:\n')
df_filled = df_missing.fillna(df_missing.median())
print (df_filled.isnull().sum())
```

<hr>
**5. Sorting and ranking**
<hr>

```python
top_salary = df.sort_values('salary', ascending=False).head(5)
print ('top 5 salaries:\n%s\n'%str(top_salary))
df['rank'] = df['salary'].rank(ascending=False)
print (df[['name','salary','rank']].head(5))
```

<hr>
**6. Basic Visualizations**
<hr>

```python
plt.figure(figsize=(10,5))
df.boxplot(column='salary', by='dept')
plt.title('Salary distribution by department')
plt.suptitle('')
plt.show()
print ('boxplot displayed\n')
```

```python
dept_avg = df.groupby('dept')['salary'].mean().sort_values()
plt.figure(figsize=(8,5))
dept_avg.plot(kind='barh', color=['blue','green','red','orange'])
plt.xlabel('Average Salary')
plt.title('Average Salary by Department')
plt.show()
print ('bar chart displayed\n')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['experience'], df['salary'], alpha=0.6, c=df['score'], cmap='viridis')
plt.colorbar(label='Score')
plt.xlabel('Experience (years)')
plt.ylabel('Salary')
plt.title('Salary vs Experience (colored by Score)')
plt.show()
print ('scatter plot displayed\n')
```

```python
plt.figure(figsize=(8,5))
df['score'].hist(bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Score Distribution')
plt.show()
print ('histogram displayed\n')
```

```python
dept_count = df['dept'].value_counts()
plt.figure(figsize=(7,7))
plt.pie(dept_count, labels=dept_count.index, autopct='%1.1f%%', startangle=90)
plt.title('Department Distribution')
plt.show()
print ('pie chart displayed\n')
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
df.groupby('dept')['salary'].mean().plot(kind='bar')
plt.title('Mean Salary')
plt.xticks(rotation=45)
plt.subplot(1,2,2)
df.groupby('dept')['score'].mean().plot(kind='bar', color='green')
plt.title('Mean Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print ('subplots displayed\n')
```
