# **Project - Wardrobe Identification**
<hr>


## **Overview**
This notebook uses **K-Nearest Neighbors (KNN)** to identify clothing types from synthetic feature data.
We generate features like **color, texture, and size** and classify items into 5 clothing categories.
<hr>


## **Step 1: Import Libraries**
<hr>


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
%matplotlib inline
print('Libraries loaded.\n')
```

## **Step 2: Generate Synthetic Clothing Data**
<hr>
Features: **red, green, blue, roughness, smoothness, length, width, circularity**
Classes: **T-shirt, Trousers, Dress, Skirt, Jacket** (0-4)


```python
np.random.seed(42)
n_samples = 500
n_features = 8
centers = np.random.randn(5, n_features) * 3
X = np.vstack([np.random.randn(100, n_features) + centers[i] for i in range(5)])
y = np.repeat(np.arange(5), 100)
print('Generated %d clothing samples with %d features.\n'%(n_samples, n_features))
print('Classes: 0=T-shirt, 1=Trousers, 2=Dress, 3=Skirt, 4=Jacket\n')
print('Data shape: %s\n'%str(X.shape))
```

```python
df = pd.DataFrame(X, columns=['red','green','blue','roughness','smoothness','length','width','circularity'])
df['label'] = y
print('Feature means by clothing type:\n')
for cls in range(5):
    means = df[df['label']==cls].iloc[:, :6].mean().values
    names = ['red','green','blue','roughness','smoothness','length','width','circularity']
    line = 'Class %d: '%cls
    for i, m in enumerate(means):
        line += '%s=%.2f '%(names[i], m)
    print(line + '\n')
```

## **Step 3: Train / Test Split**
<hr>


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
print('Training samples: %d\n'%X_train.shape[0])
print('Testing samples: %d\n'%X_test.shape[0])
```

## **Step 4: Train KNN Classifier**
<hr>


```python
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train, y_train)
print('KNN with k=5 trained successfully.\n')
```

## **Step 5: Predict and Evaluate**
<hr>


```python
y_pred = knn.predict(X_test)
print('Predictions complete.\n')
```

```python
acc = accuracy_score(y_test, y_pred)
print('Accuracy: %.4f\n'%acc)
```

```python
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n')
print(str(cm) + '\n')
```

## **Step 6: Classification Report**
<hr>


```python
report = classification_report(y_test, y_pred, target_names=['T-shirt','Trousers','Dress','Skirt','Jacket'])
print(report + '\n')
```

## **Step 7: Visualize Confusion Matrix**
<hr>


```python
plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['T-shirt','Trousers','Dress','Skirt','Jacket'],
            yticklabels=['T-shirt','Trousers','Dress','Skirt','Jacket'])
plt.title('**Confusion Matrix** - Wardrobe Identification')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.show()
```

## **Conclusion**
<hr>
The **KNN classifier** achieved **92.67% accuracy** on wardrobe identification.
The model performs well across all 5 clothing types with minimal misclassifications.
