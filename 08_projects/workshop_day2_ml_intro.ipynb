**1. Dictionaries**
<hr>

```python
student = {
    'name': 'Alice',
    'age': 22,
    'courses': ['Math', 'Physics', 'CS'],
    'grades': {'Math': 90, 'Physics': 85, 'CS': 95}
}
print ('student: %s\n'%str(student))
print ('name: %s\n'%student['name'])
print ('courses: %s\n'%str(student['courses']))
print ('CS grade: %d\n'%student['grades']['CS'])
student['age'] = 23
student['email'] = 'alice@example.com'
print ('updated: %s\n'%str(student))
```

```python
print ('keys: %s\n'%str(list(student.keys())))
print ('values: %s\n'%str(list(student.values())))
print ('items:\n')
for k, v in student.items():
    print ('  %s: %s'%(k, str(v)))
print ('\n')
print ('get phone: %s\n'%str(student.get('phone', 'NOT FOUND')))
```

**2. Sets**
<hr>

```python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}
print ('a: %s\n'%str(a))
print ('b: %s\n'%str(b))
print ('union: %s\n'%str(a | b))
print ('intersection: %s\n'%str(a & b))
print ('difference a-b: %s\n'%str(a - b))
print ('symmetric diff: %s\n'%str(a ^ b))
a.add(9)
a.discard(2)
print ('a after add/discard: %s\n'%str(a))
```

```python
text = 'hello world hello python world hello'
words = text.split()
unique_words = set(words)
print ('words: %s\n'%str(words))
print ('unique: %s\n'%str(unique_words))
print ('unique count: %d\n'%len(unique_words))
print ('is subset: {1,2} subset of {1,2,3}: %s\n'%str({1,2}.issubset({1,2,3})))
```

**3. List comprehensions**
<hr>

```python
nums = range(1, 21)
squares = [n**2 for n in nums]
evens = [n for n in nums if n % 2 == 0]
odd_squares = [n**2 for n in nums if n % 2 == 1]
print ('squares: %s\n'%str(squares))
print ('evens: %s\n'%str(evens))
print ('odd_squares: %s\n'%str(odd_squares))
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flattened = [num for row in matrix for num in row]
print ('matrix: %s\n'%str(matrix))
print ('flattened: %s\n'%str(flattened))

pairs = [(x, y) for x in [1,2,3] for y in ['a','b']]
print ('pairs: %s\n'%str(pairs))

words = ['hello', 'world', 'python', 'list']
word_lengths = {w: len(w) for w in words}
print ('word_lengths: %s\n'%str(word_lengths))
```

<hr>
**4. Lambda functions**
<hr>

```python
square = lambda x: x ** 2
add = lambda a, b: a + b
max_val = lambda a, b: a if a > b else b
print ('square(7): %d\n'%square(7))
print ('add(10, 20): %d\n'%add(10, 20))
print ('max_val(50, 30): %d\n'%max_val(50, 30))
```

```python
nums = [5, 2, 8, 1, 9, 3]
sorted_nums = sorted(nums)
sorted_desc = sorted(nums, reverse=True)
sorted_by_mod = sorted(nums, key=lambda x: x % 3)
print ('original: %s\n'%str(nums))
print ('sorted: %s\n'%str(sorted_nums))
print ('descending: %s\n'%str(sorted_desc))
print ('sorted by %%3: %s\n'%str(sorted_by_mod))

people = [('Alice', 25), ('Bob', 20), ('Charlie', 30)]
sorted_people = sorted(people, key=lambda p: p[1])
print ('sorted by age: %s\n'%str(sorted_people))
```

```python
print ('map with lambda:\n')
mapped = list(map(lambda x: x * 2, [1,2,3,4]))
print ('double: %s\n'%str(mapped))

print ('filter with lambda:\n')
filtered = list(filter(lambda x: x > 2, [1,2,3,4,5]))
print ('>2: %s\n'%str(filtered))

print ('reduce with lambda:\n')
from functools import reduce
product = reduce(lambda a, b: a * b, [1,2,3,4,5])
print ('product: %d\n'%product)
```

**5. Tuple unpacking and zip**
<hr>

```python
point = (3, 7)
x, y = point
print ('x=%d, y=%d\n'%(x, y))

names = ['Alice', 'Bob', 'Charlie']
scores = [88, 92, 75]
for name, score in zip(names, scores):
    print ('%s: %d'%(name, score))
print ('\n')
zipped = list(zip(names, scores))
print ('zipped: %s\n'%str(zipped))
unzipped_names, unzipped_scores = zip(*zipped)
print ('unzipped: %s, %s\n'%str((unzipped_names, unzipped_scores)))
```

**6. Error handling**
<hr>

```python
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return 'cannot divide by zero'
    except TypeError:
        return 'invalid type'
    finally:
        print ('division attempted')

print ('10/2 = %s\n'%str(safe_divide(10, 2)))
print ('10/0 = %s\n'%str(safe_divide(10, 0)))
print ('10/\'a\' = %s\n'%str(safe_divide(10, 'a')))
```

```python
# Word frequency counter
text = 'the quick brown fox jumps over the lazy dog the fox jumps'
word_counts = {}
for w in text.split():
    word_counts[w] = word_counts.get(w, 0) + 1
print ('word frequencies: %s\n'%str(word_counts))

# Top 3 most common
sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
print ('top 3: %s\n'%str(sorted_words[:3]))
```

```python
# Find common elements in two lists using sets
list1 = [1, 2, 3, 4, 5, 6]
list2 = [4, 5, 6, 7, 8, 9]
common = list(set(list1) & set(list2))
print ('common: %s\n'%str(common))

# Palindrome check using lambda
is_palindrome = lambda s: s == s[::-1]
print ('radar: %s\n'%str(is_palindrome('radar')))
print ('python: %s\n'%str(is_palindrome('python')))
```
