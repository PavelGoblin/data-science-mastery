# Column Transformer Example
---
<hr>
**Demonstrating sklearn's ColumnTransformer for mixed data types**

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler
print ('imports ready\n')
```

```python
np.random.seed(123)
df = pd.DataFrame({
    'age': np.random.randint(18, 80, 200),
    'salary': np.random.normal(55000, 20000, 200),
    'department': np.random.choice(['IT', 'HR', 'Sales', 'Marketing'], 200),
    'location': np.random.choice(['US', 'EU', 'Asia'], 200)
})
print ('Sample data created, shape: %s' % str(df.shape))
```

```python
df.head(8)
```

<hr>
## **Define ColumnTransformer**

```python
numeric_features = ['age', 'salary']
categorical_features = ['department', 'location']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features)
    ])
print ('ColumnTransformer defined\n')
```

```python
X = df.copy()
X_transformed = preprocessor.fit_transform(X)
print ('Before shape: %s' % str(X.shape))
print ('After shape: %s' % str(X_transformed.shape))
```

```python
print ('Feature count change: %d -> %d' % (X.shape[1], X_transformed.shape[1]))
print ('Numeric scaled, categorical one-hot encoded')
```

<hr>
## **Inspect Transformed Output**

```python
cat_encoder = preprocessor.named_transformers_['cat']
cat_feature_names = cat_encoder.get_feature_names_out(categorical_features)
all_features = np.concatenate([numeric_features, cat_feature_names])
print ('Column names after transform: %s' % str(all_features))
```

```python
print ('First 5 rows of transformed data:\n%s' % str(np.round(X_transformed[:5], 3)))
```

```python
print ('Feature means (should be ~0 for scaler):')
print (np.round(X_transformed[:, :2].mean(axis=0), 6))
```

<hr>
**ColumnTransformer** cleanly applies different transforms to different columns.
No manual splitting needed -- one fit/transform call handles everything.

```python
print ('ColumnTransformer example completed successfully.\n')
```
