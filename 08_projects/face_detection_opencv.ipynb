# Case Study: Face Detection using OpenCV

Detect faces in images using OpenCV's Haar Cascade classifiers. This is a classic computer vision application.


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

print('OpenCV version:', cv2.__version__)
```

## 1. Load Haar Cascade Classifier


```python
# Load pre-trained face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

if face_cascade.empty():
    print('Error loading cascade classifier')
else:
    print('Cascade classifier loaded successfully')
```

## 2. Helper Function for Detection


```python
def detect_faces(image_path, scale_factor=1.1, min_neighbors=5):
    """Detect faces in an image and return annotated image."""
    img = cv2.imread(str(image_path))
    if img is None:
        print(f'Could not read image: {image_path}')
        return None, 0
    
    # Convert to grayscale (Haar cascades work on grayscale)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=scale_factor, minNeighbors=min_neighbors)
    
    # Draw rectangles around faces
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    
    # Convert BGR to RGB for matplotlib display
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img_rgb, len(faces)

print('Helper function defined')
```

## 3. Test on Sample Image


```python
# Using a sample image if available, otherwise creating a test pattern
sample_path = Path('Data/kids.jpg')

if sample_path.exists():
    img, n_faces = detect_faces(sample_path)
    if img is not None:
        plt.figure(figsize=(10, 8))
        plt.imshow(img)
        plt.title(f'Detected {n_faces} face(s)')
        plt.axis('off')
        plt.show()
else:
    print(f'Sample image not found at {sample_path}')
    print('To test face detection, place an image in the Data/ folder and run:')
    print('  img, n = detect_faces("Data/your_image.jpg")')
```

## 4. Parameter Tuning


```python
print('''
Key Parameters for detectMultiScale():

- scaleFactor: How much the image size is reduced at each scale.
   1.05 = slower but more accurate, 1.3 = faster but may miss faces.

- minNeighbors: Higher values = fewer false positives.
   3-6 is typical.

- minSize: Minimum face size to detect (prevents tiny false positives).

Tip: For group photos, use scaleFactor=1.05 to detect smaller faces.
''')
```

## 5. Real-time Webcam Detection (Optional)


```python
print('''
To run real-time face detection with webcam:

  cap = cv2.VideoCapture(0)
  while True:
      ret, frame = cap.read()
      gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
      faces = face_cascade.detectMultiScale(gray, 1.1, 5)
      for (x, y, w, h) in faces:
          cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
      cv2.imshow('Face Detection', frame)
      if cv2.waitKey(1) & 0xFF == ord('q'):
          break
  cap.release()
  cv2.destroyAllWindows()
''')
```
