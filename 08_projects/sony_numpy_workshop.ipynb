<hr>
**Arrays, reshaping, broadcasting, indexing**
<hr>

```python
import numpy as np
print ('numpy version: %s\n'%np.__version__)
```

<hr>
**1. Creating arrays**
<hr>

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.zeros((2,3))
arr3 = np.ones((3,2))
arr4 = np.arange(0, 10, 2)
arr5 = np.linspace(0, 1, 5)
print ('arr1: %s\n'%str(arr1))
print ('arr2:\n%s\n'%str(arr2))
print ('arr3:\n%s\n'%str(arr3))
print ('arr4: %s\n'%str(arr4))
print ('arr5: %s\n'%str(arr5))
```

```python
arr_random = np.random.randn(3, 4)
arr_int = np.random.randint(0, 100, size=(2, 5))
print ('random normal:\n%s\n'%str(arr_random))
print ('random int:\n%s\n'%str(arr_int))
```

<hr>
**2. Array attributes**
<hr>

```python
a = np.random.randn(4, 5, 3)
print ('shape: %s\n'%str(a.shape))
print ('ndim: %d\n'%a.ndim)
print ('size: %d\n'%a.size)
print ('dtype: %s\n'%str(a.dtype))
```

<hr>
**3. Reshaping**
<hr>

```python
b = np.arange(12)
print ('original: %s\n'%str(b))
print ('reshape (3,4):\n%s\n'%str(b.reshape(3,4)))
print ('reshape (2,2,3):\n%s\n'%str(b.reshape(2,2,3)))
print ('reshape (-1,6):\n%s\n'%str(b.reshape(-1,6)))
```

```python
c = np.array([[1,2,3],[4,5,6]])
print ('original:\n%s\n'%str(c))
print ('flatten: %s\n'%str(c.flatten()))
print ('ravel: %s\n'%str(c.ravel()))
print ('T (transpose):\n%s\n'%str(c.T))
```

<hr>
**4. Indexing and slicing**
<hr>

```python
d = np.arange(16).reshape(4,4)
print ('array:\n%s\n'%str(d))
print ('d[0]: %s\n'%str(d[0]))
print ('d[1,2]: %s\n'%str(d[1,2]))
print ('d[:2]:\n%s\n'%str(d[:2]))
print ('d[:,1:3]:\n%s\n'%str(d[:,1:3]))
print ('d[::2, ::2]:\n%s\n'%str(d[::2, ::2]))
```

```python
e = np.array([10,20,30,40,50])
indices = [0,2,4]
print ('fancy indexing: %s\n'%str(e[indices]))
bool_idx = e > 25
print ('boolean mask: %s\n'%str(bool_idx))
print ('filtered: %s\n'%str(e[bool_idx]))
```

<hr>
**5. Broadcasting**
<hr>

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([10,20,30])
print ('a:\n%s\n'%str(a))
print ('b: %s\n'%str(b))
print ('a + b:\n%s\n'%str(a + b))
print ('a * 2:\n%s\n'%str(a * 2))
print ('a ** 2:\n%s\n'%str(a ** 2))
```

```python
x = np.array([[0],[1],[2]])
y = np.array([1,2,3])
print ('x shape: %s, y shape: %s\n'%(str(x.shape), str(y.shape)))
print ('x + y (broadcast):\n%s\n'%str(x + y))
```

<hr>
**6. Universal functions**
<hr>

```python
vals = np.array([0.0, np.pi/4, np.pi/2, np.pi])
print ('vals: %s\n'%str(vals))
print ('sin: %s\n'%str(np.round(np.sin(vals), 4)))
print ('cos: %s\n'%str(np.round(np.cos(vals), 4)))
print ('exp: %s\n'%str(np.round(np.exp(vals), 4)))
print ('log: %s\n'%str(np.round(np.log(vals+1), 4)))
```

<hr>
**7. Aggregations**
<hr>

```python
data = np.random.randn(4, 5)
print ('data:\n%s\n'%str(np.round(data, 3)))
print ('sum: %.4f\n'%np.sum(data))
print ('mean: %.4f\n'%np.mean(data))
print ('std: %.4f\n'%np.std(data))
print ('min: %.4f, max: %.4f\n'%(np.min(data), np.max(data)))
print ('axis=0 sum: %s\n'%str(np.round(np.sum(data, axis=0), 4)))
print ('axis=1 sum: %s\n'%str(np.round(np.sum(data, axis=1), 4)))
```

<hr>
**8. Linear algebra**
<hr>

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print ('A @ B (matmul):\n%s\n'%str(A @ B))
print ('np.dot(A, B):\n%s\n'%str(np.dot(A, B)))
print ('A * B (element-wise):\n%s\n'%str(A * B))
print ('det: %.4f\n'%np.linalg.det(A))
print ('inv:\n%s\n'%str(np.round(np.linalg.inv(A), 4)))
```

<hr>
**9. Stacking and splitting**
<hr>

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
print ('vstack:\n%s\n'%str(np.vstack([x,y])))
print ('hstack: %s\n'%str(np.hstack([x,y])))
arr = np.arange(12).reshape(3,4)
print ('split:\n')
for part in np.split(arr, 3):
    print (part)
```
