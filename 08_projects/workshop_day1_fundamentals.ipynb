**1. Variables and data types**
<hr>

```python
x = 10
y = 3.14
name = 'Python'
is_fun = True
print ('x: %d, type: %s\n'%(x, type(x).__name__))
print ('y: %.2f, type: %s\n'%(y, type(y).__name__))
print ('name: %s, type: %s\n'%(name, type(name).__name__))
print ('is_fun: %s, type: %s\n'%(str(is_fun), type(is_fun).__name__))
```

```python
print ('x + y = %.2f\n'%(x + y))
print ('x // 3 = %d\n'%(x // 3))
print ('x %% 3 = %d\n'%(x % 3))
print ('x ** 2 = %d\n'%(x ** 2))
```

<hr>
**2. Lists**
<hr>

```python
fruits = ['apple', 'banana', 'cherry', 'dates']
print ('fruits: %s\n'%str(fruits))
print ('first: %s\n'%fruits[0])
print ('last: %s\n'%fruits[-1])
print ('slice [1:3]: %s\n'%str(fruits[1:3]))
fruits.append('elderberry')
print ('after append: %s\n'%str(fruits))
fruits.remove('banana')
print ('after remove: %s\n'%str(fruits))
```

```python
numbers = [3, 7, 1, 9, 4, 6]
print ('numbers: %s\n'%str(numbers))
print ('length: %d\n'%len(numbers))
print ('max: %d\n'%max(numbers))
print ('min: %d\n'%min(numbers))
print ('sum: %d\n'%sum(numbers))
numbers.sort()
print ('sorted: %s\n'%str(numbers))
```

<hr>
**3. For loops**
<hr>

```python
print ('loop over fruits:\n')
for f in fruits:
    print (' - %s'%f)
print ('\nloop with range:\n')
for i in range(5):
    print ('i = %d'%i)
print ('\nenumerate:\n')
for idx, val in enumerate(['a','b','c']):
    print ('index %d: %s'%(idx, val))
```

<hr>
**4. While loops**
<hr>

```python
count = 0
total = 0
while count < 5:
    total += count
    count += 1
print ('sum 0..4 = %d\n'%total)

i = 10
while i > 0:
    i -= 3
    print ('i = %d'%i)
print ('done\n')
```

<hr>
**5. Conditional statements**
<hr>

```python
age = 18
if age < 13:
    print ('child\n')
elif age < 20:
    print ('teenager\n')
else:
    print ('adult\n')

score = 85
grade = 'A' if score >= 80 else 'B'
print ('grade: %s\n'%grade)
```

<hr>
**6. Functions**
<hr>

```python
def square(x):
    return x * x

def greet(name, greeting='Hello'):
    return '%s, %s!'%(greeting, name)

def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)

print ('square(5) = %d\n'%square(5))
print ('greet(\'Alice\') = %s\n'%greet('Alice'))
print ('greet(\'Bob\', \'Hi\') = %s\n'%greet('Bob', 'Hi'))
print ('factorial(5) = %d\n'%factorial(5))
```

```python
def is_even(num):
    return num % 2 == 0

print ('is_even(10): %s\n'%str(is_even(10)))
print ('is_even(7): %s\n'%str(is_even(7)))

def sum_list(nums):
    total = 0
    for n in nums:
        total += n
    return total

print ('sum_list([1,2,3,4,5]) = %d\n'%sum_list([1,2,3,4,5]))
```

**7. String operations**
<hr>

```python
s = '  Python Programming  '
print ('original: "%s"\n'%s)
print ('strip: "%s"\n'%s.strip())
print ('upper: %s\n'%s.strip().upper())
print ('lower: %s\n'%s.strip().lower())
print ('split: %s\n'%str(s.strip().split()))
print ('replace: %s\n'%s.strip().replace('Python', 'Java'))
print ('starts with Py: %s\n'%str(s.strip().startswith('Py')))
print ('find Prog: %d\n'%s.strip().find('Prog'))
```
