# **Project - Reducing Colors in Image**
<hr>


## **Overview**
This notebook uses **KMeans clustering** to reduce the number of colors in an image.
We create a synthetic 200x200 RGB image and apply KMeans with **k=8, 16, 32** to quantize the colors.
<hr>


## **Step 1: Import Libraries**
<hr>


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
print('All libraries loaded.\n')
```

## **Step 2: Generate Synthetic Image**
<hr>


```python
np.random.seed(42)
img = np.random.randint(0, 256, size=(200, 200, 3), dtype=np.uint8)
print('Image shape: %s\n'%str(img.shape))
print('Data type: %s\n'%str(img.dtype))
print('Total pixels: %d\n'%img.shape[0]*img.shape[1])
```

## **Step 3: Display Original Image**
<hr>


```python
plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.title('**Original Image** (200x200)')
plt.axis('off')
plt.tight_layout()
plt.show()
```

## **Step 4: K-Means with k=8**
<hr>


```python
h, w, c = img.shape
pixels = img.reshape(-1, 3)
kmeans8 = KMeans(n_clusters=8, random_state=42, n_init=10)
labels8 = kmeans8.fit_predict(pixels)
colors8 = kmeans8.cluster_centers_.astype(np.uint8)
quantized8 = colors8[labels8].reshape(h, w, c)
print('KMeans k=8 completed.\n')
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(quantized8)
plt.title('**Quantized Image** (k=8)')
plt.axis('off')
plt.tight_layout()
plt.show()
```

## **Step 5: K-Means with k=16**
<hr>


```python
kmeans16 = KMeans(n_clusters=16, random_state=42, n_init=10)
labels16 = kmeans16.fit_predict(pixels)
colors16 = kmeans16.cluster_centers_.astype(np.uint8)
quantized16 = colors16[labels16].reshape(h, w, c)
print('KMeans k=16 completed.\n')
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(quantized16)
plt.title('**Quantized Image** (k=16)')
plt.axis('off')
plt.tight_layout()
plt.show()
```

## **Step 6: K-Means with k=32**
<hr>


```python
kmeans32 = KMeans(n_clusters=32, random_state=42, n_init=10)
labels32 = kmeans32.fit_predict(pixels)
colors32 = kmeans32.cluster_centers_.astype(np.uint8)
quantized32 = colors32[labels32].reshape(h, w, c)
print('KMeans k=32 completed.\n')
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(quantized32)
plt.title('**Quantized Image** (k=32)')
plt.axis('off')
plt.tight_layout()
plt.show()
```

## **Step 7: Compression Ratios**
<hr>
We compare **unique colors** before and after quantization.


```python
unique_orig = len(np.unique(img.reshape(-1, 3), axis=0))
print('Original unique colors: %d\n'%unique_orig)
print('k=8  -> %d unique colors, ratio: %.1f:1\n'%(8, unique_orig/8))
print('k=16 -> %d unique colors, ratio: %.1f:1\n'%(16, unique_orig/16))
print('k=32 -> %d unique colors, ratio: %.1f:1\n'%(32, unique_orig/32))
```

## **Step 8: Display Side-by-Side Comparison**
<hr>


```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].imshow(img)
axes[0, 0].set_title('**Original**')
axes[0, 0].axis('off')
axes[0, 1].imshow(quantized8)
axes[0, 1].set_title('**k=8**')
axes[0, 1].axis('off')
axes[1, 0].imshow(quantized16)
axes[1, 0].set_title('**k=16**')
axes[1, 0].axis('off')
axes[1, 1].imshow(quantized32)
axes[1, 1].set_title('**k=32**')
axes[1, 1].axis('off')
plt.tight_layout()
plt.show()
```

## **Conclusion**
<hr>
KMeans clustering effectively reduces image colors.
Higher **k** preserves more detail; lower **k** gives higher compression.
This technique is widely used in **image compression** and **style transfer**.
