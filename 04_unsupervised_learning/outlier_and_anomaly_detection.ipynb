# 13. Anomaly Detection

## 13.1 What are Anomalies?

Anomalies, also known as **outliers**, are data points that deviate significantly from the majority of the data. They can indicate rare events, fraud, equipment faults, or errors in data collection.

Anomaly detection is the process of identifying these unusual patterns. It is widely used in:
- **Fraud detection** – identifying unusual credit card transactions
- **Network security** – detecting intrusions or cyber attacks
- **Industrial monitoring** – spotting equipment failures before they happen
- **Data cleaning** – removing erroneous observations

In this notebook we explore four popular anomaly detection algorithms available in scikit-learn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.covariance import EllipticEnvelope
from sklearn.neighbors import LocalOutlierFactor
```

## 13.2 Isolation Forest

Signature: `sklearn.**IsolationForest**(*contamination=0.1*)`

Isolation Forest isolates anomalies by randomly splitting the data. Anomalies are few and different, so they require fewer splits to isolate. The algorithm returns a score for each sample; lower scores indicate anomalies.

```python
np.random.seed(42)
X_inliers = 0.3 * np.random.randn(100, 2)
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))
X = np.vstack([X_inliers, X_outliers])

iso_forest = IsolationForest(contamination=0.1, random_state=42)
y_pred_if = iso_forest.fit_predict(X)

print ('Predictions (-1 = anomaly): %s\n' % y_pred_if)
print ('Number of anomalies detected: %d\n' % np.sum(y_pred_if == -1))
```

<hr>

## 13.3 One-Class SVM

Signature: `sklearn.**OneClassSVM**(*nu=0.1*)`

One-Class SVM learns a boundary around the normal data. The `nu` parameter controls the expected proportion of outliers. Points lying outside the learned boundary are classified as anomalies.

```python
oc_svm = OneClassSVM(nu=0.1, kernel='rbf', gamma='auto')
y_pred_svm = oc_svm.fit_predict(X)

print ('Predictions (-1 = anomaly): %s\n' % y_pred_svm)
print ('Number of anomalies detected: %d\n' % np.sum(y_pred_svm == -1))
```

<hr>

## 13.4 Elliptic Envelope

Signature: `sklearn.**EllipticEnvelope**(*contamination=0.1*)`

Elliptic Envelope fits a multivariate Gaussian distribution to the data and identifies points that lie far from the mean under the estimated covariance. It works well when the data is roughly normally distributed.

```python
elliptic = EllipticEnvelope(contamination=0.1, random_state=42)
y_pred_ee = elliptic.fit_predict(X)

print ('Predictions (-1 = anomaly): %s\n' % y_pred_ee)
print ('Number of anomalies detected: %d\n' % np.sum(y_pred_ee == -1))
```

<hr>

## 13.5 Local Outlier Factor

Signature: `sklearn.**LocalOutlierFactor**(*contamination=0.1*)`

Local Outlier Factor (LOF) computes the local density deviation of a data point with respect to its neighbours. Points with substantially lower density than their neighbours are flagged as anomalies. This method works well even when the data has clusters of varying density.

```python
lof = LocalOutlierFactor(contamination=0.1)
y_pred_lof = lof.fit_predict(X)

print ('Predictions (-1 = anomaly): %s\n' % y_pred_lof)
print ('Number of anomalies detected: %d\n' % np.sum(y_pred_lof == -1))
```

<hr>

## 13.6 Comparing Methods

Let us apply all four methods on the same dataset and compare how many anomalies each technique detects.

We also print the **overlap** between the methods to see which points are consistently flagged as outliers.

```python
results = pd.DataFrame({
    'Isolation Forest': y_pred_if,
    'One-Class SVM':    y_pred_svm,
    'Elliptic Envelope': y_pred_ee,
    'Local Outlier Factor': y_pred_lof
})

anomaly_counts = (results == -1).sum()
print ('\nAnomaly counts per method:\n')
print (anomaly_counts)
print ('\n')

consensus = (results == -1).sum(axis=1)
print ('Points flagged by all 4 methods: %d\n' % np.sum(consensus == 4))
print ('Points flagged by at least 3 methods: %d\n' % np.sum(consensus >= 3))
print ('Points flagged by at least 2 methods: %d\n' % np.sum(consensus >= 2))
```
