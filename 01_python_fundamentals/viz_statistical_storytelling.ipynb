```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style='whitegrid')
print('Seaborn version:', sns.__version__)
```

```python
tips = sns.load_dataset('tips')
print('Tips dataset:')
print(tips.head())
print(f'\nShape: {tips.shape}')
```

## 1. Distribution Analysis

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.histplot(tips['total_bill'], kde=True, bins=30, ax=axes[0,0])
axes[0,0].set_title('Total Bill Distribution')
sns.kdeplot(tips['tip'], fill=True, ax=axes[0,1])
axes[0,1].set_title('Tip Amount (KDE)')
sns.scatterplot(data=tips, x='total_bill', y='tip', alpha=0.6, ax=axes[1,0])
axes[1,0].set_title('Total Bill vs Tip')
sns.ecdfplot(tips['total_bill'], ax=axes[1,1])
axes[1,1].set_title('Cumulative Distribution')
plt.tight_layout()
plt.show()
```

## 2. Categorical Comparisons

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0,0])
axes[0,0].set_title('Total Bill by Day')
sns.violinplot(data=tips, x='time', y='tip', hue='sex', split=True, ax=axes[0,1])
axes[0,1].set_title('Tip by Time (split by gender)')
sns.barplot(data=tips, x='day', y='total_bill', hue='smoker', ax=axes[1,0])
axes[1,0].set_title('Bill by Day (smoker/non-smoker)')
sns.pointplot(data=tips, x='day', y='tip', hue='sex', dodge=True, ax=axes[1,1])
axes[1,1].set_title('Tip by Day and Gender')
plt.tight_layout()
plt.show()
```

## 3. Regression and Faceted Plots

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'alpha':0.5}, ax=axes[0])
axes[0].set_title('Tip vs Total Bill (with Regression)')
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', col='time', height=4, aspect=1.2)
plt.suptitle('Tip by Gender and Time', y=1.05)
plt.tight_layout()
plt.show()
```

## 4. FacetGrid and PairGrid

```python
g = sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True, height=4)
g.map(sns.scatterplot, 'total_bill', 'tip', alpha=0.6)
g.add_legend()
g.fig.suptitle('Faceted by Time and Smoking Status', y=1.05)
plt.show()
```

```python
g = sns.PairGrid(tips, vars=['total_bill','tip','size'], hue='time', diag_sharey=False)
g.map_upper(sns.scatterplot, alpha=0.6)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
g.add_legend()
g.fig.suptitle('PairGrid: Multi-Variable Relationships', y=1.02)
plt.show()
```
