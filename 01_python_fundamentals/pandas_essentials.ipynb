### 1. Introduction to Pandas

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.

- **Series**: One-dimensional labeled array capable of holding any data type
- **DataFrame**: Two-dimensional labeled data structure with columns of potentially different types
- **Data I/O**: Read/write CSV, Excel, SQL, JSON, HTML, Parquet and many more formats
- **Data Cleaning**: Handle missing data, duplicates, outliers with ease
- **Data Transformation**: Group, pivot, merge, join, reshape your data
- **Time Series**: Powerful date/time functionality for temporal data
- **Performance**: Built on NumPy, vectorized operations for blazing speed

Think of Pandas as **Excel on steroids** - but programmable, reproducible, and way more powerful!

### 4. Connecting Databases

Pandas can read directly from SQL databases. Let's use SQLite as an example!

<hr>


```python
# Connecting to a SQLite database
import sqlite3

# Create an in-memory database
conn = sqlite3.connect(':memory:')
cur = conn.cursor()
cur.execute("CREATE TABLE employees (id INT, name TEXT, dept TEXT, salary INT)")
cur.execute("INSERT INTO employees VALUES (1, 'Alice', 'Engineering', 85000)")
cur.execute("INSERT INTO employees VALUES (2, 'Bob', 'Marketing', 72000)")
cur.execute("INSERT INTO employees VALUES (3, 'Charlie', 'Engineering', 95000)")
conn.commit()

# Read directly into a DataFrame
df_sql = pd.read_sql_query("SELECT * FROM employees", conn)
print(df_sql)
conn.close()
```

### 7. Handling Missing Data

Real-world data is never clean. Let's learn how to handle those pesky NaN values!

<hr>


```python
# Let's create data with missing values
missing_df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [1, 2, 3, 4, 5],
    'D': [np.nan, np.nan, np.nan, np.nan, 5]
})
print("Data with missing values:")
print(missing_df)
print("\nisnull() - where are the NaNs?")
missing_df.isnull()
```

```python
# Handling missing data
missing_df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [1, 2, 3, 4, 5],
    'D': [np.nan, np.nan, np.nan, np.nan, 5]
})

# notnull()
print("notnull():")
print(missing_df.notnull())

# dropna() - drop rows with any NaN
print("\ndropna() - drop rows with any NaN:")
print(missing_df.dropna())

# fillna() - fill NaN with a value
print("\nfillna(0):")
print(missing_df.fillna(0))

# Fill with column mean
print("\nfillna with column mean:")
print(missing_df.fillna(missing_df.mean()))
```

### 8. Dropping Rows & Columns

Sometimes you just need to get rid of stuff!

<hr>


```python
# Dropping rows and columns
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500],
    'D': [1000, 2000, 3000, 4000, 5000]
})
print("Original DataFrame:")
print(df)

# Drop columns (axis=1)
print("\nDrop column B:")
print(df.drop('B', axis=1))

# Drop rows (axis=0)
print("\nDrop row 2:")
print(df.drop(2, axis=0))

# Drop multiple columns
print("\nDrop columns A and D:")
print(df.drop(['A', 'D'], axis=1))

# inplace=True modifies the original
print("\nUsing inplace=True (no new object created):")
```

### 9. Handling Duplicates

Find and remove duplicate rows in your dataset.

<hr>


```python
# Finding and removing duplicates
dup_df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob', 'Diana'],
    'age': [25, 30, 35, 25, 30, 28],
    'city': ['NYC', 'LA', 'Chicago', 'NYC', 'LA', 'Boston']
})
print("Original with duplicates:")
print(dup_df)

# duplicated() - marks duplicate rows
print("\nduplicated():")
print(dup_df.duplicated())

# drop_duplicates() - removes duplicate rows
print("\ndrop_duplicates():")
print(dup_df.drop_duplicates())

# Keep only the last occurrence
print("\ndrop_duplicates(keep='last'):")
print(dup_df.drop_duplicates(keep='last'))

# Check duplicates in specific columns only
print("\nduplicated() on name column only:")
print(dup_df.duplicated(subset=['name']))
```

### 10. Function Application

Transform your data with map(), apply(), groupby(), rolling() and string methods!

<hr>


```python
# map() - apply a function element-wise to a Series
df = pd.DataFrame({'name': ['Alice', 'Bob', 'Charlie'],
                   'score': [85, 92, 78]})

# map with a dictionary
grade_map = {85: 'B', 92: 'A', 78: 'C'}
df['grade'] = df['score'].map(grade_map)
print("Using map() with dictionary:")
print(df)

# apply() - apply a function along an axis
print("\nUsing apply() on columns (axis=0):")
print(df[['score']].apply(lambda x: x * 1.1))

# apply() with a custom function
def categorize(score):
    if score >= 90:
        return 'Excellent'
    elif score >= 80:
        return 'Good'
    else:
        return 'Needs Improvement'

df['category'] = df['score'].apply(categorize)
print("\nUsing apply() with custom function:")
print(df)
```

```python
# groupby() - split-apply-combine
url = "https://raw.githubusercontent.com/zekelabs/data-science-complete-tutorial/master/Data/cars.csv"
cars = pd.read_csv(url)

# Group by 'Origin' and calculate mean MPG
print("Average MPG by Origin:")
print(cars.groupby('Origin')['MPG'].mean())

# Group by multiple columns
print("\nAverage MPG by Origin and Cylinders:")
print(cars.groupby(['Origin', 'Cylinders'])['MPG'].mean())

# rolling() - rolling window calculations
print("\nRolling mean (window=3) of first 10 MPG values:")
print(cars['MPG'].head(10).rolling(window=3).mean())
```

```python
# String methods - vectorized string operations
data = pd.DataFrame({'text': ['hello world', 'PYTHON is FUN', '   pandas   ', 'DATA SCIENCE']})

# str.lower()
data['lower'] = data['text'].str.lower()
# str.upper()
data['upper'] = data['text'].str.upper()
# str.strip() - remove whitespace
data['stripped'] = data['text'].str.strip()
# str.contains() - check if string contains a pattern
data['has_pandas'] = data['text'].str.contains('pandas', case=False)
# str.replace()
data['replaced'] = data['text'].str.replace(' ', '_')
# str.split()
data['split'] = data['text'].str.split()

print("String operations:")
data
```

### 11. Merge, Join & Concatenate

Combine multiple DataFrames like a pro!

<hr>


```python
# Merge, Join & Concatenate
# merge() - like SQL JOIN
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'id': [1, 2, 4], 'score': [85, 92, 78]})

# Inner merge (only matching ids)
print("Inner merge:")
print(pd.merge(df1, df2, on='id', how='inner'))

# Left merge (keep all from df1)
print("\nLeft merge:")
print(pd.merge(df1, df2, on='id', how='left'))

# concat() - stack DataFrames vertically or horizontally
df_a = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df_b = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

print("\nVertical concatenation (concat along axis=0):")
print(pd.concat([df_a, df_b], axis=0))

print("\nHorizontal concatenation (concat along axis=1):")
print(pd.concat([df_a, df_b], axis=1))

# join() - join on index
df_x = pd.DataFrame({'salary': [50000, 60000]}, index=['Alice', 'Bob'])
df_y = pd.DataFrame({'age': [25, 30]}, index=['Alice', 'Bob'])
print("\nJoin on index:")
print(df_x.join(df_y))
```

### 12. Pivot Tables

Summarize your data with pivot tables - the Excel feature, now in Python!

<hr>


```python
# pivot_table() - create a spreadsheet-style pivot table
url = "https://raw.githubusercontent.com/zekelabs/data-science-complete-tutorial/master/Data/cars.csv"
cars = pd.read_csv(url)

# Create a pivot table: average MPG by Origin and Cylinders
pivot = pd.pivot_table(cars,
                       values='MPG',
                       index='Origin',
                       columns='Cylinders',
                       aggfunc='mean')
print("Pivot Table - Average MPG by Origin and Cylinders:")
pivot
```

### 13. Normalizing JSON

Flatten nested JSON into a flat table!

<hr>


```python
# json_normalize() - flatten nested JSON
from pandas import json_normalize

# Nested JSON data (list of dicts with nested dicts)
data = [
    {
        'name': 'Alice',
        'age': 30,
        'address': {'city': 'New York', 'state': 'NY', 'zip': 10001},
        'hobbies': ['reading', 'cycling']
    },
    {
        'name': 'Bob',
        'age': 25,
        'address': {'city': 'Los Angeles', 'state': 'CA', 'zip': 90001},
        'hobbies': ['gaming', 'hiking']
    },
    {
        'name': 'Charlie',
        'age': 35,
        'address': {'city': 'Chicago', 'state': 'IL', 'zip': 60601},
        'hobbies': ['cooking', 'photography']
    }
]

# Flatten the nested JSON
flat = json_normalize(data, sep='_')
print("Normalized JSON (nested address flattened):")
flat
```
