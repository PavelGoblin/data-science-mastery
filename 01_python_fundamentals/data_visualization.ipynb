# Plotting with Python

A picture is worth a thousand words — and in data science, a good plot is worth a thousand tables. Let's learn how to bring your data to life!

We'll cover:
- **Matplotlib** — the OG plotting library
- **Pandas plotting** — quick plots from DataFrames
- **Seaborn** — beautiful statistical plots with minimal code
- **Subplots, bars, histograms, box plots, and more!**

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.datasets import load_iris, load_digits, load_wine

print("Libraries loaded successfully! 🎉")
```

## 1. Basic Plotting with Matplotlib

Matplotlib is the foundation of Python plotting. Think of it as your blank canvas — you add lines, labels, and colors to create the picture you want.

<hr>

```python
# Simple line plot — our first chart!
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# Scatter plot — great for seeing relationships between two variables
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5

plt.scatter(x, y, color='purple', alpha=0.6)
plt.show()
```

```python
# Customized plot with labels, title, and legend
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label='sin(x)', color='blue', linewidth=2)
plt.plot(x, np.cos(x), label='cos(x)', color='red', linestyle='--')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Sine and Cosine Waves')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 2. Plotting with Pandas

Pandas has built-in plotting that wraps Matplotlib. If your data is in a DataFrame, you can plot with one line of code!

<hr>

```python
# Create a DataFrame and plot it directly
df = pd.DataFrame({
    'Year': [2018, 2019, 2020, 2021, 2022, 2023],
    'Sales': [150, 180, 200, 220, 280, 350],
    'Costs': [120, 140, 160, 170, 190, 210]
})

df.plot(x='Year', y=['Sales', 'Costs'], kind='line', marker='o')
plt.title('Company Performance Over Time')
plt.ylabel('Amount ($)')
plt.show()
```

```python
# Bar plot from a DataFrame — perfect for comparing categories
df_scores = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Score': [88, 92, 79, 95, 84]
})

df_scores.plot(x='Student', y='Score', kind='bar', color='teal', legend=False)
plt.title('Student Exam Scores')
plt.ylabel('Score')
plt.ylim(0, 100)
plt.show()
```

```python
# Area plot — shows volume over time nicely
df.plot(x='Year', y=['Sales', 'Costs'], kind='area', alpha=0.5)
plt.title('Sales vs Costs (Area View)')
plt.ylabel('Amount ($)')
plt.show()
```

## 3. Subplots

Sometimes one plot isn't enough! Subplots let you arrange multiple charts in a grid.

<hr>

```python
# Using plt.subplots() — returns figure and array of axes
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(0, 10, 100)

axes[0, 0].plot(x, np.sin(x), color='blue')
axes[0, 0].set_title('sin(x)')

axes[0, 1].plot(x, np.cos(x), color='red')
axes[0, 1].set_title('cos(x)')

axes[1, 0].plot(x, np.tan(x), color='green')
axes[1, 0].set_ylim(-5, 5)
axes[1, 0].set_title('tan(x)')

axes[1, 1].plot(x, np.exp(-x) * np.sin(x), color='purple')
axes[1, 1].set_title('Damped oscillation')

plt.tight_layout()
plt.show()
```

```python
# Subplots with different plot types
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Line plot
axes[0].plot(iris_df.iloc[:50, 0])
axes[0].set_title('Sepal length (first 50)')

# Histogram
axes[1].hist(iris_df.iloc[:, 0], bins=15, color='skyblue', edgecolor='black')
axes[1].set_title('Sepal length distribution')

# Scatter
axes[2].scatter(iris_df.iloc[:, 0], iris_df.iloc[:, 1], alpha=0.6, c=iris.target)
axes[2].set_xlabel('Sepal length')
axes[2].set_ylabel('Sepal width')
axes[2].set_title('Sepal length vs width')

plt.tight_layout()
plt.show()
```

## 4. Bar Charts & Histograms

Bar charts = categories. Histograms = distributions. Both are essential for exploratory data analysis.

<hr>

```python
# Bar chart — comparing categories
categories = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
values = [25, 40, 15, 30, 20]
colors = ['#FF6B6B', '#FFD93D', '#FF8E9E', '#6BCB77', '#4D96FF']

plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.2)
plt.title('Fruit Inventory')
plt.ylabel('Quantity')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Histogram — shows the distribution of a numerical variable
np.random.seed(42)
heights = np.random.normal(170, 10, 1000)

plt.hist(heights, bins=30, color='lightcoral', edgecolor='black', alpha=0.8)
plt.title('Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram with multiple groups (stacked)
data_a = np.random.normal(60, 10, 500)
data_b = np.random.normal(75, 12, 500)

plt.hist([data_a, data_b], bins=20, label=['Group A', 'Group B'],
         color=['skyblue', 'salmon'], alpha=0.7, edgecolor='black')
plt.title('Comparing Two Groups')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## 5. Box Plots & Violin Plots

Box plots show the five-number summary (min, Q1, median, Q3, max). Violin plots add the density shape on top. Both are great for spotting outliers and comparing distributions.

<hr>

```python
# Box plot — understand the spread of your data
np.random.seed(100)
data = [np.random.normal(50, 10, 100) for _ in range(4)]

plt.boxplot(data, labels=['Model A', 'Model B', 'Model C', 'Model D'],
            patch_artist=True,
            boxprops=dict(facecolor='lightblue'))
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy (%)')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Violin plot — like a box plot but shows the full distribution shape
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Using matplotlib's violin plot
axes[0].violinplot(data, showmeans=True, showmedians=True)
axes[0].set_xticks([1, 2, 3, 4])
axes[0].set_xticklabels(['A', 'B', 'C', 'D'])
axes[0].set_title('Violin Plot')
axes[0].grid(axis='y', alpha=0.3)

# Box plot side by side for comparison
axes[1].boxplot(data, labels=['A', 'B', 'C', 'D'], patch_artist=True,
                boxprops=dict(facecolor='lightgreen'))
axes[1].set_title('Box Plot')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Using Seaborn

Seaborn makes **beautiful** statistical plots with minimal code. It works great with Pandas DataFrames and has built-in themes that look professional out of the box.

<hr>

```python
# Load the Iris dataset and create a pairplot — the EDA superstar!
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

sns.set_style('whitegrid')
sns.pairplot(iris_df, hue='species', palette='Set2', height=2.5)
plt.suptitle('Iris Dataset — Pairplot', y=1.02)
plt.show()
```

```python
# Heatmap — visualize correlations
corr = iris_df.iloc[:, :-1].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={'shrink': 0.75})
plt.title('Iris Feature Correlations')
plt.show()
```

```python
# Seaborn box plot — much prettier than plain matplotlib!
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=iris_df, palette='Set3')
plt.title('Sepal Length by Species')
plt.show()
```

```python
# Seaborn violin plot + swarm plot combo
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.violinplot(x='species', y='petal length (cm)', data=iris_df, palette='muted')
plt.title('Petal Length — Violin Plot')

plt.subplot(1, 2, 2)
sns.swarmplot(x='species', y='petal length (cm)', data=iris_df, palette='dark')
plt.title('Petal Length — Swarm Plot')

plt.tight_layout()
plt.show()
```

```python
# Distribution plot with Seaborn
np.random.seed(42)
exam_scores = np.random.normal(72, 15, 500)

sns.histplot(exam_scores, kde=True, bins=25, color='steelblue')
plt.title('Exam Score Distribution')
plt.xlabel('Score')
plt.axvline(exam_scores.mean(), color='red', linestyle='--', label=f'Mean = {exam_scores.mean():.1f}')
plt.legend()
plt.show()
```

## 7. Advanced Customizations

Let's make our plots pop! Colors, markers, line styles, figure sizes, grids — the little details that turn a good plot into a great one.

<hr>

```python
# Colors, markers, and line styles — the full palette
x = np.linspace(0, 5, 20)

plt.figure(figsize=(10, 6))

plt.plot(x, x**1.5, 'ro-', label='x^1.5', linewidth=2, markersize=8)
plt.plot(x, x**2, 'bs--', label='x^2', linewidth=2, markersize=8)
plt.plot(x, x**2.5, 'g^:', label='x^2.5', linewidth=2, markersize=8)
plt.plot(x, x**3, 'm*-.', label='x^3', linewidth=2, markersize=8)

plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.title('Power Functions with Different Styles', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.4, linestyle=':')
plt.tight_layout()
plt.show()
```

```python
# Figure size, axis limits, and text annotations
np.random.seed(42)
x = np.random.randn(50)
y = 2 * x + np.random.randn(50) * 0.3

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=x, cmap='viridis', s=80, alpha=0.7, edgecolors='black', linewidth=0.5)
plt.colorbar(label='X value')

plt.xlim(-3, 3)
plt.ylim(-6, 6)
plt.xlabel('Feature X', fontsize=12)
plt.ylabel('Target Y', fontsize=12)
plt.title('Colored Scatter Plot with Colorbar', fontsize=13, fontweight='bold')

# Add an annotation
plt.annotate('Outlier!', xy=(2.5, 5.2), xytext=(1.5, 4.5),
             arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=11, color='red')

plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
# Customizing Seaborn themes
sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1.2)

# Load wine dataset
wine = load_wine()
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df['target'] = wine.target

# Create a styled joint plot
g = sns.jointplot(data=wine_df, x='alcohol', y='malic_acid', hue='target',
                  palette='Set1', kind='scatter', height=7)
g.fig.suptitle('Alcohol vs Malic Acid by Wine Class', y=1.02, fontweight='bold')
plt.show()

# Reset to default style for remaining cells
sns.set_theme()
```

```python
# Putting it all together — a polished, publication-ready figure
np.random.seed(2024)

fig, ax = plt.subplots(figsize=(10, 6))

categories = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = [120, 190, 175, 240]
profit = [30, 55, 45, 80]

x_pos = np.arange(len(categories))
width = 0.35

bars1 = ax.bar(x_pos - width/2, revenue, width, label='Revenue',
               color='#2196F3', edgecolor='white', linewidth=0.8)
bars2 = ax.bar(x_pos + width/2, profit, width, label='Profit',
               color='#4CAF50', edgecolor='white', linewidth=0.8)

ax.set_xlabel('Quarter', fontsize=12, fontweight='bold')
ax.set_ylabel('Amount ($K)', fontsize=12, fontweight='bold')
ax.set_title('Annual Performance by Quarter', fontsize=14, fontweight='bold')
ax.set_xticks(x_pos)
ax.set_xticklabels(categories)
ax.legend(frameon=True, fancybox=True, shadow=True)
ax.grid(axis='y', alpha=0.3, linestyle='--')

# Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords='offset points', ha='center', fontsize=9)
for bar in bars2:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords='offset points', ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Bonus: Plotting the digits dataset — visualizing images!
digits = load_digits()

fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for i, ax in enumerate(axes.ravel()):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f'Label: {digits.target[i]}')
    ax.axis('off')

plt.suptitle('Sample Digits from sklearn Dataset', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
