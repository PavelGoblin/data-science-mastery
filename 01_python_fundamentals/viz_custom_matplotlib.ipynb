```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
print('Matplotlib version:', mpl.__version__)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
colors = ['#2E86AB','#A23B72','#F18F01','#C73E1D','#3B1F2B']
mpl.rcParams['figure.dpi'] = 120
print('Custom style applied')
```

## 1. Dual-Axis Plot

```python
x = np.linspace(0, 10, 100)
fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.plot(x, np.sin(x), color=colors[0], linewidth=2.5, label='sin(x)')
ax1.set_ylabel('sin(x)', color=colors[0], fontsize=13)
ax1.tick_params(axis='y', labelcolor=colors[0])
ax1.legend(loc='upper left')
ax2 = ax1.twinx()
ax2.plot(x, np.cos(x), color=colors[2], linewidth=2.5, label='cos(x)')
ax2.set_ylabel('cos(x)', color=colors[2], fontsize=13)
ax2.tick_params(axis='y', labelcolor=colors[2])
ax2.legend(loc='upper right')
plt.title('Dual-Axis Plot', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 2. Custom Annotations

```python
x = np.arange(1, 13)
y = 50 + 30*np.sin(2*np.pi*x/12) + np.random.normal(0, 5, 12)
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(x, y, 'o-', color=colors[0], linewidth=2, markersize=8, zorder=3)
ax.fill_between(x, y, alpha=0.2, color=colors[0])
max_idx, min_idx = np.argmax(y), np.argmin(y)
ax.annotate(f'Peak: {y[max_idx]:.1f}', xy=(x[max_idx], y[max_idx]),
            xytext=(x[max_idx]+1.5, y[max_idx]+5),
            arrowprops=dict(facecolor=colors[3], shrink=0.05, width=2),
            fontweight='bold', color=colors[3])
ax.annotate(f'Trough: {y[min_idx]:.1f}', xy=(x[min_idx], y[min_idx]),
            xytext=(x[min_idx]-1.5, y[min_idx]-8),
            arrowprops=dict(facecolor=colors[4], shrink=0.05, width=2),
            fontweight='bold', color=colors[4])
ax.set_title('Annotated Time Series', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3. Custom Colormap Heatmap

```python
from matplotlib.colors import LinearSegmentedColormap
cmap = LinearSegmentedColormap.from_list('custom', ['#2166AC','#F7F7F7','#B2182B'], N=256)
data = np.random.randn(10, 10)
data = (data + data.T) / 2
np.fill_diagonal(data, 1)
fig, ax = plt.subplots(figsize=(9, 7))
im = ax.imshow(data, cmap=cmap, vmin=-1, vmax=1, aspect='equal')
for i in range(10):
    for j in range(10):
        ax.text(j, i, f'{data[i,j]:.2f}', ha='center', va='center',
                color='white' if abs(data[i,j]) > 0.5 else 'black', fontsize=8)
plt.colorbar(im, ax=ax, shrink=0.7, label='Correlation')
ax.set_title('Custom Heatmap', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 4. GridSpec Multi-Panel Layout

```python
fig = plt.figure(figsize=(15, 10))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1, 1:])
ax4 = fig.add_subplot(gs[2, :2])
ax5 = fig.add_subplot(gs[2, 2])
x = np.linspace(0, 2*np.pi, 100)
ax1.plot(x, np.sin(2*x), color=colors[0], linewidth=2)
ax1.set_title('sin(2x) - Wide Top')
ax2.plot(x, np.cos(2*x), color=colors[1], linewidth=2)
ax2.set_title('cos(2x)')
ax3.bar([1,2,3,4], [3,5,2,7], color=colors[2])
ax3.set_title('Bar Panel')
ax4.scatter(x, np.sin(3*x)*np.exp(-x/5), c=x, cmap='viridis', s=20, alpha=0.6)
ax4.set_title('Damped Scatter')
ax5.pie([35,25,20,20], labels=['A','B','C','D'], colors=colors[:4], autopct='%1.0f%%')
ax5.set_title('Pie Chart')
fig.suptitle('GridSpec Multi-Panel Layout', fontsize=15, fontweight='bold', y=1.01)
plt.show()
```
