## 1. Variables & Data Types


```python
# 1. Integer
a = 10; b = -5; c = 0; d = 0xFF; e = 0b1010; f = 1_000_000
print(f'Int: {a}, {b}, {c}, hex={d}, binary={e}, underscore={f}')
print(f'Type: {type(a)}, Size: {a.bit_length()} bits')
print(f'Max int: {sys.maxsize}')
```

```python
# 2. Float
x = 3.14; y = -0.001; z = 1.5e10; w = float('inf'); v = float('nan')
print(f'Float: {x}, {y}, {z}, inf={w}, nan={v}')
print(f'Is NaN: {v != v}, Is inf: {w == float("inf")}')
print(f'Round: {round(x, 1)}, Floor: {int(x)}, Ceil: {int(x)+1}')
```

```python
# 3. Complex numbers
c1 = 3 + 4j; c2 = complex(1, -2)
print(f'Complex: {c1}, Real={c1.real}, Imag={c1.imag}, Conj={c1.conjugate()}')
print(f'Magnitude: {abs(c1)}, Add: {c1 + c2}, Mul: {c1 * c2}')
```

```python
# 4. Boolean
t = True; f = False
print(f'Bool: {t}, {f}, Int: {int(t)}, {int(f)}')
print(f'AND: {t and f}, OR: {t or f}, NOT: {not t}, XOR: {t != f}')
print(f'Truthy: {bool(1)}, {bool(-1)}, {bool("hello")}, {bool([1,2])}')
print(f'Falsy: {bool(0)}, {bool("")}, {bool([])}, {bool(None)}')
```

```python
# 5. NoneType
n = None
print(f'None: {n}, Type: {type(n)}')
print(f'Is None: {n is None}, Is not: {n is not None}')
def func(): return None
r = func(); print(f'Returns: {r}')
```

```python
# 6. Type conversion
print(f'int to float: {float(10)}')
print(f'float to int: {int(3.99)}')
print(f'str to int: {int("42")}, str to float: {float("3.14")}')
print(f'int to str: {str(100)}, float to str: {str(3.14)}')
print(f'list to tuple: {tuple([1,2,3])}, tuple to list: {list((1,2,3))}')
print(f'to bool: {bool(0)}, {bool("")}, {bool([1])}')
try: int('abc')
except: print('int("abc") -> ValueError')
```

```python
# 7. Type checking
print(f'isinstance(10, int): {isinstance(10, int)}')
print(f'isinstance(3.14, (int, float)): {isinstance(3.14, (int, float))}')
print(f'type(10) == int: {type(10) == int}')
print(f'issubclass(bool, int): {issubclass(bool, int)}')
print(f'callable(print): {callable(print)}, callable(10): {callable(10)}')
```

```python
# 8. Multiple assignment & swapping
x, y, z = 1, 2, 3; print(f'Multi assign: {x}, {y}, {z}')
x, y = y, x; print(f'Swap: x={x}, y={y}')
a = b = c = 0; print(f'Chain assign: {a}, {b}, {c}')
*head, tail = [1,2,3,4]; print(f'Star unpack: head={head}, tail={tail}')
first, *middle, last = [1,2,3,4,5]; print(f'First={first}, Middle={middle}, Last={last}')
```

```python
# 9. Variable scope
global_var = 'global'
def scope_test():
    local_var = 'local'
    print(f'Inside: {local_var}, Global: {global_var}')
scope_test()
# print(local_var)  # would raise NameError

count = 0
def increment():
    global count; count += 1
increment(); increment(); print(f'Global count: {count}')
```

```python
# 10. Constants & naming conventions
PI = 3.14159  # convention: uppercase = constant
MAX_RETRIES = 3
_private = 'underscore prefix = internal use'
__mangled = 'double underscore triggers name mangling'
print(f'PI={PI}, MAX_RETRIES={MAX_RETRIES}, private={_private}')
print(f'Mangled: {__mangled}')
```

```python
# 8. List copy
l = [1, 2, 3]
l2 = l; print(f'Reference: l2 is l: {l2 is l}')
l3 = l.copy(); print(f'Copy: l3 is l: {l3 is l}')
l4 = l[:]; print(f'Slicing copy: l4 is l: {l4 is l}')
l5 = list(l); print(f'list() copy: l5 is l: {l5 is l}')
import copy
l6 = copy.deepcopy(l); print(f'Deepcopy: {l6}')
```

```python
# 9. Nested lists
matrix = [[1, 2], [3, 4], [5, 6]]
print(f'matrix: {matrix}')
print(f'matrix[0][1]: {matrix[0][1]}')
print(f'matrix[1]: {matrix[1]}')
flat = [x for row in matrix for x in row]; print(f'flattened: {flat}')
rows, cols = 3, 4
grid = [[0] * cols for _ in range(rows)]
print(f'grid: {grid}')
# Watch out: this creates shared references!
bad = [[0] * cols] * rows; bad[0][0] = 1; print(f'bad grid: {bad}')
```

```python
# 10. List as stack & queue
stack = []
for x in range(5): stack.append(x); print(f'push {x}: {stack}')
while stack: print(f'pop: {stack.pop()}, remaining: {stack}')

from collections import deque
queue = deque([1, 2, 3])
queue.append(4); print(f'enqueue: {queue}')
print(f'dequeue: {queue.popleft()}, queue: {queue}')
```

```python
# 11. List comprehensions — advanced
nums = [1, 2, 3, 4, 5]
print(f'squares: {[x**2 for x in nums]}')
print(f'evens: {[x for x in nums if x % 2 == 0]}')
print(f'conditional: {["even" if x%2==0 else "odd" for x in nums]}')
print(f'cartesian: {[(x,y) for x in [1,2] for y in [3,4]]}')
print(f'flatten: {[x for row in [[1,2],[3,4]] for x in row]}')
```

```python
# 12. List iteration techniques
l = ['a', 'b', 'c', 'd']
for i, v in enumerate(l): print(f'  index={i}, value={v}')
for v in reversed(l): print(f'  reversed: {v}', end=' ')
print('')
for v in sorted(l, reverse=True): print(f'  sorted desc: {v}', end=' ')
```

```python
# 13. List performance (timeit)
import timeit
setup = 'l = list(range(1000))'
print(f'append: {timeit.timeit("l.append(1)", setup, number=100000):.4f}s')
print(f'insert front: {timeit.timeit("l.insert(0,1)", setup, number=10000):.4f}s')
print(f'pop end: {timeit.timeit("l.pop()", setup, number=100000):.4f}s')
print(f'pop front: {timeit.timeit("l.pop(0)", setup, number=10000):.4f}s')
```

## 14. Exception Handling


```python
# 1. try/except/else/finally
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print('Cannot divide by zero!')
    except TypeError as e:
        print(f'Type error: {e}')
    else:
        print(f'Result: {result}')
    finally:
        print('Cleanup done')
divide(10, 2); print('---'); divide(10, 0); print('---'); divide(10, 'a')
```

```python
# 3. Custom exceptions
class ValidationError(Exception): pass
class NegativeValueError(ValidationError): pass

def validate_age(age):
    if age < 0: raise NegativeValueError(f'Negative: {age}')
    if age > 150: raise ValidationError(f'Too old: {age}')
    return f'Valid: {age}'
for v in [25, -5, 200]:
    try: print(validate_age(v))
    except NegativeValueError as e: print(f'Negative: {e}')
    except ValidationError as e: print(f'Invalid: {e}')
```

## 18. Decorators


```python
# 1. Basic decorator
def timer(func):
    import time
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(f'{func.__name__}: {time.time()-start:.4f}s')
        return result
    return wrapper
@timer
def slow_sum(n): return sum(range(n))
print(f'sum={slow_sum(10_000_000)}')
```

```python
# 2. Decorator with args
def repeat(n):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            for _ in range(n): func(*args, **kwargs)
        return wrapper
    return decorator
@repeat(3)
def say(msg): print(f'  {msg}')
say('Hello!')
```

```python
# 3. Multiple decorators
def bold(func): return lambda: f'<b>{func()}</b>'
def italic(func): return lambda: f'<i>{func()}</i>'
@bold
@italic
def hello(): return 'Hello'
print(f'decorated: {hello()}')
```

## 20. Object-Oriented Programming


```python
# 1. Basic class
class Dog:
    species = 'Canis familiaris'
    def __init__(self, name, age):
        self.name = name; self.age = age
    def bark(self): return f'{self.name} says Woof!'
    def __str__(self): return f'{self.name}, {self.age}yrs'
d = Dog('Rex', 3)
print(f'{d}')
print(f'{d.bark()}')
print(f'species: {Dog.species}')
```

```python
# 2. Inheritance
class Puppy(Dog):
    def __init__(self, name, age, toy):
        super().__init__(name, age)
        self.toy = toy
    def bark(self): return f'{self.name} yaps! (loves {self.toy})'
p = Puppy('Max', 1, 'bone')
print(f'{p}')
print(f'{p.bark()}')
print(f'isinstance: {isinstance(p, Dog)}, {isinstance(p, Puppy)}')
```

```python
# 3. Multiple inheritance & MRO
class A:
    def m(self): return 'A'
class B(A):
    def m(self): return 'B'
class C(A):
    def m(self): return 'C'
class D(B, C):
    pass
print(f'MRO: {[c.__name__ for c in D.__mro__]}')
print(f'D().m()={D().m()}')

```

```python
# 4. Magic methods
class Vector:
    def __init__(self, x, y): self.x, self.y = x, y
    def __add__(self, v): return Vector(self.x+v.x, self.y+v.y)
    def __mul__(self, s): return Vector(self.x*s, self.y*s)
    def __abs__(self): return (self.x**2 + self.y**2)**0.5
    def __eq__(self, v): return self.x==v.x and self.y==v.y
    def __repr__(self): return f'V({self.x},{self.y})'
v1 = Vector(1, 2); v2 = Vector(3, 4)
print(f'{v1} + {v2} = {v1+v2}')
print(f'{v1} * 3 = {v1*3}')
print(f'abs({v1}) = {abs(v1):.2f}')
print(f'{v1} == {v2}: {v1==v2}')
```

```python
# 5. @property decorator
class Temperature:
    def __init__(self, celsius): self._celsius = celsius
    @property
    def celsius(self): return self._celsius
    @celsius.setter
    def celsius(self, v):
        if v < -273.15: raise ValueError('Too cold')
        self._celsius = v
    @property
    def fahrenheit(self): return self._celsius * 9/5 + 32
t = Temperature(25); print(f'C={t.celsius}, F={t.fahrenheit}')
t.celsius = 30; print(f'C={t.celsius}, F={t.fahrenheit}')
try: t.celsius = -300
except ValueError as e: print(f'Error: {e}')
```

```python
# 6. @classmethod & @staticmethod
class Date:
    def __init__(self, y, m, d): self.y, self.m, self.d = y, m, d
    @classmethod
    def from_string(cls, s):
        y, m, d = map(int, s.split('-'))
        return cls(y, m, d)
    @staticmethod
    def is_valid(s):
        try: y,m,d = map(int, s.split('-')); return True
        except: return False
    def __str__(self): return f'{self.y}-{self.m:02d}-{self.d:02d}'
d = Date.from_string('2024-01-15')
print(f'from_string: {d}')
print(f'is_valid: {Date.is_valid("2024-01-15")}')
```

```python
# 7. Abstract base classes
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self): pass
    @abstractmethod
    def perimeter(self): pass
class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return 3.14159 * self.r ** 2
    def perimeter(self): return 2 * 3.14159 * self.r
c = Circle(5); print(f'Circle area={c.area():.2f}, perim={c.perimeter():.2f}')
```

```python
# 8. Dataclasses (3.7+)
from dataclasses import dataclass, field
@dataclass(order=True)
class Person:
    name: str
    age: int
    city: str = 'Unknown'
    tags: list = field(default_factory=list)
p1 = Person('Alice', 30); p2 = Person('Bob', 25, 'NYC')
print(f'{p1}')
print(f'{p2}')
print(f'p1 == Person("Alice", 30): {p1 == Person("Alice", 30)}')
print(f'sorted by age: {sorted([p1,p2])}')
```

```python
import re
text = 'Contact: alice@email.com, bob@test.org, or call 555-1234'
emails = re.findall(r'[\w.]+@[\w.]+', text)
print(f'emails: {emails}')
phone = re.search(r'\d{3}-\d{4}', text)
print(f'phone: {phone.group() if phone else "none"}')
print(f'sub: {re.sub(r"\d{3}-\d{4}", "[REDACTED]", text)}')
print(f'split: {re.split(r"[, ]+", text)[:5]}')
```

```python
# Regex groups & flags
log = 'ERROR 2024-01-15: Disk full'
m = re.match(r'(\w+)\s+(\d{4}-\d{2}-\d{2}):\s+(.+)', log)
if m: print(f'level={m.group(1)}, date={m.group(2)}, msg={m.group(3)}')
text2 = 'Hello\nWorld\nHello'
print(f'multiline: {re.findall(r"^Hello", text2, re.MULTILINE)}')
print(f'ignorecase: {re.findall(r"hello", "Hello HELLO hello", re.IGNORECASE)}')
```

## 26. collections — Advanced Data Structures


```python
from collections import namedtuple, deque, ChainMap, UserDict, UserList
# namedtuple
Point = namedtuple('Point', ['x', 'y'])
p = Point(10, 20); print(f'Point: {p}, x={p.x}, y={p.y}, unpack: {list(p)}')
# ChainMap
d1 = {'a': 1, 'b': 2}; d2 = {'b': 3, 'c': 4}
cm = ChainMap(d1, d2); print(f'ChainMap: {cm["a"]}, {cm["b"]}, {cm["c"]}')
# deque
dq = deque([1,2,3], maxlen=5)
dq.appendleft(0); dq.append(4); print(f'deque: {dq}')
```

## 27. Threading & Concurrency


```python
import threading, time
results = []
def worker(n):
    time.sleep(0.1); results.append(n*n)
threads = [threading.Thread(target=worker, args=(i,)) for i in range(5)]
for t in threads: t.start()
for t in threads: t.join()
print(f'thread results: {sorted(results)}')
print(f'active threads: {threading.active_count()}')
```

## 29. Performance & Best Practices


```python
import timeit
# List vs generator
print(f'list comprehension: {timeit.timeit("[x**2 for x in range(1000)]", number=10000):.3f}s')
print(f'generator: {timeit.timeit("(x**2 for x in range(1000))", number=10000):.3f}s')
# local variable lookup is faster
def test():
    s = 0
    for i in range(1000): s += i
    return s
print(f'test(): {timeit.timeit(test, number=10000):.3f}s')
```

## 30. Common Gotchas & Pitfalls


```python
# Mutable default args
def append_to(item, lst=[]):
    lst.append(item); return lst
print(f'gotcha: {append_to(1)}, {append_to(2)}, {append_to(3)}')
# Fix: use None
def append_to_fixed(item, lst=None):
    if lst is None: lst = []
    lst.append(item); return lst
print(f'fixed: {append_to_fixed(1)}, {append_to_fixed(2)}, {append_to_fixed(3)}')
```

```python
# Late binding closures
funcs = [lambda: i for i in range(5)]
print(f'late bind: {[f() for f in funcs]}')
# Fix: default arg
funcs2 = [lambda i=i: i for i in range(5)]
print(f'fixed: {[f() for f in funcs2]}')
```

```python
# is vs ==
a = [1, 2, 3]; b = [1, 2, 3]
print(f'a == b: {a == b}, a is b: {a is b}')
c = a
print(f'a == c: {a == c}, a is c: {a is c}')
# small integer caching
x, y = 256, 256; print(f'x=256 is y=256: {x is y}')
x, y = 257, 257; print(f'x=257 is y=257: {x is y}')
```
